# Support Vector Machines

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

np.random.seed(0)

n_points = 1000

data1 = 0.5 * np.random.randn(n_points, 2) + [0, 0]
data2 = 0.5 * np.random.randn(n_points, 2) + [2, 2]

X = np.vstack([data1, data2])
y = np.hstack([np.zeros(n_points), np.ones(n_points)])

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolors='k')
plt.title('Synthetic 2D Data for SVM')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

clf = SVC(kernel='poly')
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.4f}%")

# Plotting the decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='rainbow', s=50, edgecolors='k')

#get x and y values to run through decision function to get boundaries 
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx = np.linspace(xlim[0], xlim[1])
yy = np.linspace(ylim[0], ylim[1])
YY, XX = np.meshgrid(yy, xx)

xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

plt.title('SVM Decision Boundary')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

clf = SVC(kernel='linear')
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.4f}%")

# Plotting the decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='rainbow', s=50, edgecolors='k')

#get x and y values to run through decision function to get boundaries 
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx = np.linspace(xlim[0], xlim[1])
yy = np.linspace(ylim[0], ylim[1])

#mesh to produce gutters of decision boundaries 
YY, XX = np.meshgrid(yy, xx)

#get each 
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

plt.title('SVM Decision Boundary')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
