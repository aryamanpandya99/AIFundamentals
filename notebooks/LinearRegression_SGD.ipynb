```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def LinearRegression(epochs, L, x, y, k):
    n = x.size
    m_hat = 0
    c_hat = 0 
    y_hat = np.full((len(y)), 0)
    two_norm = []
    iteration = []

    #run for number of epochs specified above, with the given learning rate L 
    for epoch in range(epochs): 
        iteration.append(epoch)
        idx = np.random.choice(np.arange(len(x)), k, replace=False)
        x_sample = x[idx]
        y_sample = y[idx]
        dc = 0
        dm = 0 
            #making a prediction for y based on current m and c hat values 
        y_hat = (m_hat * x)+c_hat
            #calculating the partial derivatives for both m and c hat values 
        dm = (-2/n)*sum(x_sample*(y_sample-y_hat[idx]))
        dc = (-2/n)*sum(y_sample-y_hat[idx])
        two_norm.append(np.linalg.norm((y_sample-y_hat[idx]), ord = 2))
            #correct m and c hat values based on gradient descent 
        m_hat = m_hat - L * (dm)
        c_hat = c_hat - L * (dc)
    #plotting the same scatter plot as previously displayed 
    fig1, ax1 = plt.subplots()
    ax1.scatter(x, y)
    ax1.set_ylabel('Y')
    ax1.set_xlabel('X')
    #plotting the regression line obtained above 
    ax1.plot([min(x), max(x)], [min(y_hat), max(y_hat)], color='black') 
    
    fig2, ax2 = plt.subplots()
    ax2.set_ylabel('Error')
    ax2.set_xlabel('Iteration')
    ax2.plot(iteration, two_norm)
```

```python
m = np.full((100), 5)
c = np.full((100), 1)
x = np.random.uniform(-5,5,[100])
noise = np.random.normal(size = 100)
```

```python
y = (m*x)+c+noise 
plt.scatter(x, y)
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```

```python
def getLearningRate(current_rate, decay_rate, epoch):
    return (1/(1+decay_rate*epoch))*current_rate
```

```python
def LinearRegressionDecaying(epochs, L, x, y, k):
    n = x.size
    m_hat = 0
    c_hat = 0 
    y_hat = np.full((1,len(y)), 0)
    two_norm = []
    iteration = []

    #run for number of epochs specified above, with the given learning rate L 
    for epoch in range(epochs): 
        iteration.append(epoch)
        idx = np.random.choice(np.arange(len(x)), k, replace=False)
        x_sample = x[idx]
        y_sample = y[idx]
        dc = 0
        dm = 0 
            #making a prediction for y based on current m and c hat values 
        y_hat = (m_hat * x)+c_hat
            #calculating the partial derivatives for both m and c hat values 
        dm = (-2/n)*sum(x_sample*(y_sample-y_hat[idx]))
        dc = (-2/n)*sum(y_sample-y_hat[idx])
        two_norm.append(np.linalg.norm(y_sample-y_hat[idx])) 
            #correct m and c hat values based on gradient descent 
        L = getLearningRate(L, 0.1, epoch)
        m_hat = m_hat - L * (dm)
        c_hat = c_hat - L * (dc)
    #plotting the same scatter plot as previously displayed 
    fig1, ax1 = plt.subplots()
    ax1.scatter(x, y)
    ax1.set_ylabel('Y')
    ax1.set_xlabel('X')
    #plotting the regression line obtained above 
    ax1.plot([min(x), max(x)], [min(y_hat), max(y_hat)], color='black') 
    
    fig2, ax2 = plt.subplots()
    ax2.set_ylabel('Error')
    ax2.set_xlabel('Iteration')
    ax2.plot(iteration, two_norm)
```

```python
learning_rate = 30
LinearRegressionDecaying(epochs, learning_rate, x, y, 10)
```
