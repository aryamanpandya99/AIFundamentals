```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def LinearRegression(epochs, L, x, y):
    n = x.size
    m_hat = 0
    c_hat = 0 
    y_hat = np.full((100), 0)

    #run for number of epochs specified above, with the given learning rate L 
    for epoch in range(epochs): 
        #making a prediction for y based on current m and c hat values 
        y_hat = (m_hat * x)+c_hat
        #calculating the partial derivatives for both m and c hat values 
        dm = (-2/n)*sum(x*(y-y_hat))
        dc = (-2/n)*sum(y-y_hat)
        #correct m and c hat values based on gradient descent 
        m_hat = m_hat - L * dm
        c_hat = c_hat - L * dc
    #plotting the same scatter plot as previously displayed 
    plt.scatter(x, y)
    plt.ylabel('Y')
    plt.xlabel('X')
    #plotting the regression line obtained above 
    plt.plot([min(x), max(x)], [min(y_hat), max(y_hat)], color='black')
    plt.show()
```

```python
m = np.full((100), 5)
c = np.full((100), 1)
x = np.random.uniform(-5,5,[100])
noise = np.random.normal(size = 100)
```

```python
y = (m*x)+c+noise 
plt.scatter(x, y)
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```

```python
epochs = 10000
learning_rate = 0.0001
LinearRegression(epochs, learning_rate, x, y)
```
